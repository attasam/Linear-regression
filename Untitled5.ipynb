```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('house.txt', names = ['Area', 'Price'])
```

```python

X_df = pd.DataFrame(data.Area)
y_df = pd.DataFrame(data.Price)
m = len(y_df)
```

```python
plt.figure(figsize=(10,8))
plt.plot(X_df, y_df, 'kx')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
iterations = 1500
alpha = 0.01
```

```python

X_df['intercept'] = 1


X = np.array(X_df)
y = np.array(y_df).flatten()
theta = np.array([0, 0])
```

```python
def cost_function(X, y, theta):

    m = len(y) 
    
    
    J = np.sum((X.dot(theta)-y)**2)/2/m
    
    return J
```

```python
cost_function(X, y, theta)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):

    cost_history = [0] * iterations
    
    for iteration in range(iterations):
        hypothesis = X.dot(theta)
        loss = hypothesis-y
        gradient = X.T.dot(loss)/m
        theta = theta - alpha*gradient
        cost = cost_function(X, y, theta)
        cost_history[iteration] = cost

    return theta, cost_history
```

```python
(t, c) = gradient_descent(X,y,theta,alpha, iterations)
```

```python
print (t)
```

```python
print (np.array([3.5, 1]).dot(t))
print (np.array([7, 1]).dot(t))
```

```python
best_fit_x = np.linspace(0, 25, 20)
best_fit_y = [t[1] + t[0]*xx for xx in best_fit_x]
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_df.Area, y_df, '.')
plt.plot(best_fit_x, best_fit_y, '-')
plt.axis([0,25,-5,25])
plt.xlabel('Area in squre feet')
plt.ylabel('Price')
plt.title('Price vs. Area with Linear Regression Line')
```
